website: https://weather.com/weather/monthly/l/f892433d7660da170347398eb8e3d722d8d362fe7dd15af16ce88324e1b96e70

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

response = requests.get("https://weather.com/weather/monthly/l/f892433d7660da170347398eb8e3d722d8d362fe7dd15af16ce88324e1b96e70")
doc = BeautifulSoup(response.text)
```

```python
# June 26 (one day)

for each in doc.find_all("button", class_="Button--default--3zkvy CalendarDateCell--dayCell--2s4Tp CalendarDateCell--firstCell--1FnOi CalendarDateCell--firstRow--1_xoa"):
    for date in each.find_all("span",class_="CalendarDateCell--date--3Fw3h"):
        date.text
    for tempmax in each.find_all("div",class_="CalendarDateCell--tempHigh--2VBba"):
        tempmax.text
    print("June",date.text,tempmax.text)
    
```

```python
# June 27 to July 2

for each in doc.find_all("button", class_="Button--default--3zkvy CalendarDateCell--dayCell--2s4Tp CalendarDateCell--firstRow--1_xoa"):
    for date in each.find_all("span",class_="CalendarDateCell--date--3Fw3h"):
        date.text
    for tempmax in each.find_all("div",class_="CalendarDateCell--tempHigh--2VBba"):
        tempmax.text
    if int(date.text) >= 27:
        # turn date.text to integers first before if statement
        print("June",date.text,tempmax.text) 
    else:
        print("July",date.text,tempmax.text) 
```

## Make a dataframe

1. First, make several small lists: nyc1, nyc2, nyc3, nyc4;
2. Second, make a new big list: nyc_weather = nyc1 + nyc2 + nyc3 + nyc4
3. Then, Turn nyc_weather into a dataframe
4. Final step, save to a new csv

```python
# June 26 (one day)

nyc1=[]

for each in doc.find_all("button", class_="Button--default--3zkvy CalendarDateCell--dayCell--2s4Tp CalendarDateCell--firstCell--1FnOi CalendarDateCell--firstRow--1_xoa"):
    for date in each.find_all("span",class_="CalendarDateCell--date--3Fw3h"):
        date.text
    for tempmax in each.find_all("div",class_="CalendarDateCell--tempHigh--2VBba"):
        tempmax.text
    #print("June",date.text,tempmax.text)
    
    weather={
        "date": date.text,
        "tempmax":tempmax.text,
    }

    nyc1.append(weather)
    
print(nyc1)
    
```

```python
# June 27 to July 2
nyc2=[]

for each in doc.find_all("button", class_="Button--default--3zkvy CalendarDateCell--dayCell--2s4Tp CalendarDateCell--firstRow--1_xoa"):
    for date in each.find_all("span",class_="CalendarDateCell--date--3Fw3h"):
        date.text
    for tempmax in each.find_all("div",class_="CalendarDateCell--tempHigh--2VBba"):
        tempmax.text
    
    weather={
        "date": date.text,
        "tempmax":tempmax.text,
    }

    nyc2.append(weather)

print(nyc2)
   
```

```python
# July 3, July 10, July 17, July 24
nyc3=[]

for each in doc.find_all("button", class_="Button--default--3zkvy CalendarDateCell--dayCell--2s4Tp CalendarDateCell--firstCell--1FnOi"):
    for date in each.find_all("span",class_="CalendarDateCell--date--3Fw3h"):
        date.text
    for tempmax in each.find_all("div",class_="CalendarDateCell--tempHigh--2VBba"):
        tempmax.text
        
    weather ={
        "date": date.text,
        "tempmax":tempmax.text,
    }

    nyc3.append(weather)
print(nyc3)
```

```python
# July 4 to July 30
nyc4=[]

for each in doc.find_all("button", class_="Button--default--3zkvy CalendarDateCell--dayCell--2s4Tp"):
    for date in each.find_all("span",class_="CalendarDateCell--date--3Fw3h"):
        date.text
    for tempmax in each.find_all("div",class_="CalendarDateCell--tempHigh--2VBba"):
        tempmax.text
        
    weather ={
        "date": date.text,
        "tempmax":tempmax.text,
    }

    nyc4.append(weather)
print(nyc4)
```

```python
nyc_weather = nyc1+nyc2+nyc3+nyc4
nyc_weather

df = pd.DataFrame(nyc_weather)
df
```

```python
# Drop the values from June!!
# axis=0 means drop selected rows
# axis=1 means drop selected columns
df_July = df.drop(axis=0,labels=[0,1,2,3,4])
df_July.date = df_July.date.astype(int)
df_July
```

```python
df_July = df_July.sort_values(by ='date')
df_July.to_csv("July_2022_tempmax.csv")
```

# Clean the saved csv file!

```python
import pandas as pd
df_July2 = pd.read_csv("July_2022_tempmax.csv")
```

```python
# Clean the column values a little bit!
df_July2.tempmax = df_July2.tempmax.str.replace("°","")
```

# Convert date columns to datetime

Python strftime cheatsheet: https://strftime.org/

```python
# Fill the date column value with zero using str.zfill function
df_July2.date = df_July2.date.astype(str)
df_July2.date = df_July2.date.str.zfill(2)

# Create new columns named "Month" and "Year"
# Combine Year & Month & date columns into a new column named "DATE"
df_July2["Year"] = "2022"
df_July2["Month"] = "07"
df_July2["Date"] = df_July2.Year + df_July2.Month + df_July2.date.astype(str)
```

```python
df_July2["DATE"] = pd.to_datetime(df_July2.Date, format = "%Y%m%d",errors='coerce')
df_July2
```
