```python
import pandas as pd
df = pd.read_csv("Cool_It__NYC_2020_-_Cooling_Sites.csv")
df.shape
```

```python
df = df.dropna(subset=["PropertyName"])
df = df.dropna(subset=["Status"])
df.shape
```

```python
df.FeatureType.value_counts(dropna=False)
df
```

```python
df_SS = df[df.FeatureType != "Misting Station"]
df_SS.shape
```

```python
df_SS.groupby(by="Status").PropertyName.value_counts().to_csv("spray_showers_after_2020_map.csv")
```

```python
df_MS = df[df.FeatureType == "Misting Station"]
df_MS
```

```python
df_MS.groupby(by="Status").PropertyName.value_counts().to_csv("misting_station_map.csv")
```
