```python
import pandas as pd
df = pd.read_csv("Cool_It__NYC_2020_-_Drinking_Fountains.csv")
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
# Drop rows that has NaN values on selected columns
df = df.dropna(subset=["DF_Activated"])
df = df.dropna(subset=["PropertyName"])
df = df.dropna(subset=["Filter_Installed"])
```

```python
df
```

```python
# Outdoor Drinking Fountains: 3326
df.Outdoor.value_counts()
```

```python
df.DF_Activated.value_counts()
```

```python
# PropertyName: Name of the property the drinking fountain is within
df.PropertyName.value_counts()
```

```python
df.groupby(by="DF_Activated").PropertyName.value_counts().to_csv("drinking_fountains_map2.csv")
```

```python
# Whether installed a filter: 
df.Filter_Installed.value_counts()
```

```python
# Whether painted or not: 
df.Painted.value_counts(dropna=False)
```
